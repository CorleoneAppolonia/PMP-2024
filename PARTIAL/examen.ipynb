```python
#1.1:
import numpy as np
import pandas as pd
from hmmlearn import hmm
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)


df = pd.read_csv('grid_culori.csv', header=None)
grid_culori = df.to_numpy()


culori = [
    "red", "blue", "green", "yellow",
    "purple", "orange", "pink", "cyan",
    "brown", "lime"
]


culoare_to_idx = {culoare: idx for idx, culoare in enumerate(culori)}
idx_to_culoare = {idx: culoare for idx, culoare in enumerate(culori)}

observatii = ["red", "red", "lime", "yellow", "blue"]
observatii_idx = [culoare_to_idx[culoare] for culoare in observatii]
observatii_idx_np = np.array(observatii_idx).reshape(-1, 1)

dimensiune_grid = (10, 10)


stari_ascunse = [(i, j) for i in range(dimensiune_grid[0]) for j in range(dimensiune_grid[1])]
numar_stari = len(stari_ascunse)
stare_to_idx = {stare: idx for idx, stare in enumerate(stari_ascunse)}
idx_to_stare = {idx: stare for idx, stare in enumerate(stari_ascunse)}

# Matrice de tranziție cu posibilitatea de a rămâne pe loc
transitions = np.zeros((numar_stari, numar_stari))

for i, j in stari_ascunse:
    idx_current = stare_to_idx[(i, j)]
    vecini = []
    if i > 0:
        vecini.append((i - 1, j))  # sus
    if i < dimensiune_grid[0] - 1:
        vecini.append((i + 1, j))  # jos
    if j > 0:
        vecini.append((i, j - 1))  # stanga
    if j < dimensiune_grid[1] - 1:
        vecini.append((i, j + 1))  # dreapta

    numar_vecini = len(vecini)
    probabilitate_miscare = 0.75 / numar_vecini if numar_vecini > 0 else 0
    # Probabilitatea de a ramane pe loc
    transitions[idx_current, idx_current] = 0.25
    # Probabilitățile de a se muta la vecinii valizi
    for v in vecini:
        idx_vecin = stare_to_idx[v]
        transitions[idx_current, idx_vecin] = probabilitate_miscare

# Matrice de emisie
emissions = np.zeros((numar_stari, len(culori)))

for idx_state, (i, j) in enumerate(stari_ascunse):
    culoare_stare = grid_culori[i, j]
    idx_culoare = culoare_to_idx[culoare_stare]
    emissions[idx_state, idx_culoare] = 1.0

# Modelul HMM
model = hmm.CategoricalHMM(n_components=numar_stari, n_iter=100, init_params='')
model.startprob_ = np.full(numar_stari, 1.0 / numar_stari)
model.transmat_ = transitions
model.emissionprob_ = emissions

#  algoritmul Viterbi
logprob, secventa_stari = model.decode(observatii_idx_np, algorithm="viterbi")


drum = [idx_to_stare[idx] for idx in secventa_stari]

print("Cea mai probabilă secvență de poziții:")
for idx, (i, j) in enumerate(drum):
    print(f"Pasul {idx + 1}: Poziția ({i}, {j}), Culoarea '{grid_culori[i, j]}'")
print(f"\nLog-probabilitatea drumului: {logprob}")

# Vizualizarea drumului pe grid
fig, ax = plt.subplots(figsize=(8, 8))
for i in range(dimensiune_grid[0]):
    for j in range(dimensiune_grid[1]):
        culoare = grid_culori[i, j]
        ax.add_patch(plt.Rectangle((j, dimensiune_grid[0] - i - 1), 1, 1, color=culoare))

# evidentiez drumul
for idx, (i, j) in enumerate(drum):
    ax.add_patch(plt.Circle((j + 0.5, dimensiune_grid[0] - i - 0.5), 0.3, color="black", alpha=0.7))
    ax.text(j + 0.5, dimensiune_grid[0] - i - 0.5, str(idx + 1),
            color="white", ha="center", va="center", fontsize=10, fontweight="bold")

ax.set_xlim(0, dimensiune_grid[1])
ax.set_ylim(0, dimensiune_grid[0])
ax.set_xticks([])
ax.set_yticks([])
ax.grid(visible=True, color="black", linewidth=0.5)
ax.set_aspect("equal")
plt.title("Drumul rezultat al stărilor ascunse", fontsize=14)
plt.show()
```

```python
#1.2
# A doua cerinta: fara posibilitatea de a ramane pe loc
transitions_no_stay = np.zeros((numar_stari, numar_stari))

for i, j in stari_ascunse:
    idx_current = stare_to_idx[(i, j)]
    vecini = []
    if i > 0:
        vecini.append((i - 1, j))  # sus
    if i < dimensiune_grid[0] - 1:
        vecini.append((i + 1, j))  # jos
    if j > 0:
        vecini.append((i, j - 1))  # stanga
    if j < dimensiune_grid[1] - 1:
        vecini.append((i, j + 1))  # dreapta

    numar_vecini = len(vecini)
    probabilitate_miscare = 1.0 / numar_vecini if numar_vecini > 0 else 0
    # prob de a se muta la vecinii valizi
    for v in vecini:
        idx_vecin = stare_to_idx[v]
        transitions_no_stay[idx_current, idx_vecin] = probabilitate_miscare

# actualizez modelul
model_no_stay = hmm.CategoricalHMM(n_components=numar_stari, n_iter=100, init_params='')
model_no_stay.startprob_ = np.full(numar_stari, 1.0 / numar_stari)
model_no_stay.transmat_ = transitions_no_stay
model_no_stay.emissionprob_ = emissions

#  algoritmul Viterbi
logprob_ns, secventa_stari_ns = model_no_stay.decode(observatii_idx_np, algorithm="viterbi")


drum_ns = [idx_to_stare[idx] for idx in secventa_stari_ns]

print("\nCea mai probabila secventa de pozitii (fara a ramane pe loc):")
for idx, (i, j) in enumerate(drum_ns):
    print(f"Pasul {idx + 1}: Pozitia ({i}, {j}), Culoarea '{grid_culori[i, j]}'")
print(f"\nLog-probabilitatea drumului: {logprob_ns}")

# Vizualizarea noului drum pe grid
fig, ax = plt.subplots(figsize=(8, 8))
for i in range(dimensiune_grid[0]):
    for j in range(dimensiune_grid[1]):
        culoare = grid_culori[i, j]
        ax.add_patch(plt.Rectangle((j, dimensiune_grid[0] - i - 1), 1, 1, color=culoare))

#  noul drum
for idx, (i, j) in enumerate(drum_ns):
    ax.add_patch(plt.Circle((j + 0.5, dimensiune_grid[0] - i - 0.5), 0.3, color="black", alpha=0.7))
    ax.text(j + 0.5, dimensiune_grid[0] - i - 0.5, str(idx + 1),
            color="white", ha="center", va="center", fontsize=10, fontweight="bold")

ax.set_xlim(0, dimensiune_grid[1])
ax.set_ylim(0, dimensiune_grid[0])
ax.set_xticks([])
ax.set_yticks([])
ax.grid(visible=True, color="black", linewidth=0.5)
ax.set_aspect("equal")
plt.title("Drumul rezultat fara posibilitatea de a ramane pe loc", fontsize=14)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def plot_beta(alpha, beta_param, title):
    theta = np.linspace(0, 1, 1000)
    distrib = beta.pdf(theta, alpha, beta_param)
    plt.plot(theta, distrib, label=f'Beta({alpha}, {beta_param})')
    plt.title(title)
    plt.xlabel('Probabilitatea de cap (θ)')
    plt.ylabel('Densitate')
    plt.legend()
    plt.grid(True)
    plt.show()

# ========================
# distributia a priori (Beta(1,1))
# ========================

alpha_prior = 1
beta_prior = 1

plt.figure(figsize=(8, 6))
plot_beta(alpha_prior, beta_prior, 'Distributia a priori Beta(1,1)')
```

```python
# ========================
# Primul Experiment
# ========================

# Datele din primul experiment
# Secventa: s, b, b, b, s, b, s, b, b, b
# Numar de succese (cap) si escuri (pajura)
succese_exp1 = 3  # pozițiile 1, 5, 7
esecuri_exp1 = 7  # restul de 7 aruncări

# Calcularea parametrilor distribuției a posteriori
alpha_post1 = alpha_prior + succese_exp1  # 1 + 3 = 4
beta_post1 = beta_prior + esecuri_exp1    # 1 + 7 = 8

plt.figure(figsize=(8, 6))
plot_beta(alpha_post1, beta_post1, 'Distributia a posteriori dupa primul experiment')
```

```python
# ========================
# Al Doilea Experiment
# ========================

# Datele din al doilea experiment
# Secvența: s, s, b, s, b, b, s, b, b, s
# Număr de succese (cap) și eșecuri (pajură)
succese_exp2 = 5  # pozițiile 1, 2, 4, 7, 10
esecuri_exp2 = 5  # restul de 5 aruncări

# ----- Cazul a) Prior uniform (Beta(1,1)) -----

# Resetăm parametrii distribuției a priori la Beta(1,1)
alpha_prior2a = 1
beta_prior2a = 1

# Plotarea distribuției a priori pentru cazul a)
plt.figure(figsize=(8, 6))
plot_beta(alpha_prior2a, beta_prior2a, 'Distributia a priori Beta(1,1) (Cazul a)')
# Calcularea parametrilor distribuției a posteriori
alpha_post2a = alpha_prior2a + succese_exp2  # 1 + 5 = 6
beta_post2a = beta_prior2a + esecuri_exp2    # 1 + 5 = 6

# Plotarea distribuției a posteriori pentru cazul a)
plt.figure(figsize=(8, 6))
plot_beta(alpha_post2a, beta_post2a, 'Distributia a posteriori cu prior uniform (Cazul a)')
```

```python
# ========================
# Compararea distributiilor
# ========================

# Vectorul de valori pentru θ
theta = np.linspace(0, 1, 1000)

# Calcularea densitatilor ptr distrib a posteriori din cazurile a) și b)
posterior2a = beta.pdf(theta, alpha_post2a, beta_post2a)
posterior2b = beta.pdf(theta, alpha_post2b, beta_post2b)

#
plt.figure(figsize=(8, 6))
plt.plot(theta, posterior2a, label=f'Posterior Beta({alpha_post2a}, {beta_post2a}) - Prior uniform')
plt.plot(theta, posterior2b, label=f'Posterior Beta({alpha_post2b}, {beta_post2b}) - Prior informat')
plt.title('Compararea distributiilor a posteriori (Experimentul 2)')
plt.xlabel('Probabilitatea de cap (θ)')
plt.ylabel('Densitate')
plt.legend()
plt.grid(True)
plt.show()
```
